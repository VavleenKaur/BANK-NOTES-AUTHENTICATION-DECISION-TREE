```python
#VAVLEEN KAUR
# BANK-NOTES AUTHENTICATION USING DECISION TREE 

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import missingno as msn
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

data=pd.read_csv("https://raw.githubusercontent.com/VavleenKaur/BANK-NOTES-AUTHENTICATION-DECISION-TREE/main/Bank_notes.csv")
df=pd.DataFrame(data=data)

#checking missing values
df.isnull().sum()
msn.bar(df)
plt.show()

X=df.iloc[:,0:4]
Y=df["Target"]

sns.heatmap(df.corr(), annot=True)
plt.show()

```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
#C5.0
model=DecisionTreeClassifier(criterion='entropy',max_depth=3)
model.fit(X_train,Y_train)
tree.plot_tree(model)
```

```python
fn = ["variance", "skewness", "curtosis", "entropy"]
cn = ["fake","authentic"]
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(4,4), dpi=300)
tree.plot_tree(model, feature_names=fn, class_names = cn, filled=True)
```

```python
pred = model.predict(X_test)
pd.Series(pred).value_counts()
```

```python
#PREDICTING ACCURACY
print("Accuracy Scory: ",np.mean(pred==Y_test))# 0.94909
pd.crosstab(Y_test, pred)
```

```python
tn, fp, fn, tp = confusion_matrix(Y_test, pred).ravel()
accuracy = (tp+tn)/(tn+fp+fn+tp)
precision = (tp)/(tp+fp)
recall = (tp)/(tp+fn)
print(f'Accuracy is : ', accuracy)
print(f'Precision is : ', precision)
print(f'Recall is : ', recall)
```

```python
print("----------------------------------------------\n")
print("\n Confusion Matrix for Decision Tree Classifier (entropy) -\n")
print(confusion_matrix(Y_test,pred))
print("----------------------------------------------\n")
print("\nClassification Report for Decision Tree Classifier (entropy) -\n")
print(classification_report(Y_test,pred))
print("----------------------------------------------\n")
```

```python
# CART (classification and regression tree) :
# for classification :
model_gini = DecisionTreeClassifier(criterion='gini', max_depth=3)
model_gini.fit(X_train, Y_train)
gini_pred = model_gini.predict(X_test)
pd.Series(gini_pred).value_counts()
print("Accuracy Score:")
print(np.mean(Y_test==gini_pred))  
```

```python
tn, fp, fn, tp = confusion_matrix(Y_test, gini_pred).ravel()
accuracy = (tp+tn)/(tn+fp+fn+tp)
precision = (tp)/(tp+fp)
recall = (tp)/(tp+fn)
print(f'Accuracy is : ', accuracy)
print(f'Precision is : ', precision)
print(f'Recall is : ', recall)
```

```python
print("----------------------------------------------\n")
print("\n Confusion Matrix for Decision Tree Classifier (CART) -\n")
print(confusion_matrix(Y_test,gini_pred))
print("----------------------------------------------\n")
print("\nClassification Report for Decision Tree Classifier (CART) -\n")
print(classification_report(Y_test,gini_pred))
print("----------------------------------------------\n")
```

```python
reg_model = DecisionTreeRegressor()
reg_model.fit(X_train, Y_train)
reg_model.score(X_train, Y_train)          
reg_pred = reg_model.predict(X_test)
pd.crosstab(reg_pred, Y_test)
print("Accuracy Score: ")
np.mean(reg_pred==Y_test)                    
```
